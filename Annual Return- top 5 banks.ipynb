```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from pandas_datareader import data as wb
```

```python
tickers = ['TD', 'BMO', 'CM', 'RY', 'BNS']
data= pd.DataFrame()
for j in tickers:
    data[j] = wb.DataReader(j, data_source='yahoo', start= '2020-1-1')['Adj Close']
```

```python
data
```

```python
fig= ((data/data.iloc[0])*100).plot(figsize= (15,6))
plt.title('Day Return Plot', fontsize=20)
plt.xlabel('Timeframe 2020-2021', fontsize=18)
plt.ylabel('Day Return', fontsize=16)
```

```python
returns = (data / data.shift(1)) - 1
```

```python
annual = returns.mean()*250
```

```python
Return_percentage = round(annual,3)*100
```

```python
Return_percentage.plot(figsize = (15,5))
plt.title('Annual Return Plot', fontsize=20)
plt.xlabel('Timeframe 2020-2021', fontsize=18)
plt.ylabel('Annual Return %', fontsize=16)
plt.show()
```
